```python
import mne 
import os
import mypy
import glob
```

```python
dane = mne.io.read_raw_egi('CAT110 20131108 2040002.raw', preload=True)
```

```python
montage_type = 'GSN-HydroCel-65_1.0'
```

```python
montage = mne.channels.read_montage(montage_type)
from mypy.chans import correct_egi_channel_names

correct_egi_channel_names(dane)
```

```python
dane.set_montage(montage)
```

```python
dane.plot()
```

```python
dane.plot()
```

```python
%matplotlib inline
```

```python
dane.filter(0.5, None)
```

```python
din_channels = ['DIN1', 'DIN2', 'DIN4', 'DIN8', 'DI16', 'DI32', 'DI64', 'D128']
dane_din_channels = dane.copy().pick_channels(din_channels)
```

```python
zdarzenia = mypy.events.get_events_from_din(dane)
```

```python
dane.drop_channels(din_channels)
```

```python
bodźce = [36, 40, 44, 52, 56, 60]
```

```python
#epoki zmieniam na foki, aby mieć pewność, że nie kopiuję bezmyślnie z notebook'a
```

```python
foki = mne.Epochs(dane, events=zdarzenia, event_id=bodźce,
                   tmin=-0.5, tmax=0.5, preload=True)
```

```python
montage = mne.channels.read_montage(montage_type)
from mypy.chans import correct_egi_channel_names

correct_egi_channel_names(foki)
```

```python
from mne.preprocessing import ICA
```

```python
%matplotlib
```

```python
foki.plot(n_channels=64, n_epochs=3);
```

```python
# ['E34', 'E38', 'E37', 'E36', 'E19', 'E15'] zostały oznaczone jako złe, ponieważ sygnał z tych elektrod nie był dobry i wprowadzały bardzo dużo szumu
#E15 jakby "odpadła" a E19 zaczęła niepoprawnie rejestrować od połowy badania
```

```python
ica = ICA(method='extended-infomax', n_components=0.99)
```

```python
#n_components=1.0 na 1.0, a nie na 0.99 ponieważ w przeciwnnym razie pojawiał się błąd że wartość jest za mała i radzi trochę ją podnieść 
```

```python
ica.fit(foki)
```

```python
ica.plot_properties(picks=0, inst=foki, image_args=dict(sigma=2.5))
```
